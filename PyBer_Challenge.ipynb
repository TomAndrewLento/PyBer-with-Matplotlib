# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
# pyber_summary_df.rename("pyber_data_df")
# Display the data table for preview
pyber_data_df.sort_values('date')
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
city_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
city_ride_count.head()
```

```python
# 2. Get the total drivers for each city type
total_drivers_fe_city = pyber_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_fe_city
```

```python
#  3. Get the total amount of fares for each city type
total_fares_fe_city = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_fe_city
```

```python
#  4. Get the average fare per ride for each city type. 
city_type_avg_fare = pyber_data_df.groupby(["type"]).mean()["fare"]
city_type_avg_fare.head()
```

```python
# 5. Get the average fare per driver for each city type. 
city_type_driver_avg_fare=pd.Series(total_fares_fe_city/total_drivers_fe_city,name='Average Fare per Driver')
# city_type_driver_avg_fare = pyber_data_df.groupby(["type"]).sum()["fare"] / pyber_data_df.groupby(["type"]).sum()["driver_count"] 
city_type_driver_avg_fare
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df=pd.merge(city_ride_count, total_drivers_fe_city, how="left", on=["type"])
pyber_summary_df=pd.merge(pyber_summary_df, total_fares_fe_city, how="left", on=["type"])
pyber_summary_df=pd.merge(pyber_summary_df, city_type_avg_fare, how="left", on=["type"])
pyber_summary_df=pd.merge(pyber_summary_df, city_type_driver_avg_fare, how="left", on=["type"])
pyber_summary_df.rename(columns = {'ride_id':'Total Rides', 
                                   'driver_count':'Total Drivers',
                                   'fare_x':'Total Fares',
                                   'fare_y':'Average Fare per Ride',
                                   'fare':'Average Fare per Driver',
                                   }, 
                                    inplace = True)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fare_sum_by_type_date=pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
fare_sum_by_type_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fare_sum_by_type_date = fare_sum_by_type_date.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

# rural_fares = pyber_data_df.loc[(pyber_data_df["type"] == "Rural")].groupby(["date"]).sum()["fare"]
# suburban_fares = pyber_data_df.loc[(pyber_data_df["type"] == "Suburban")].groupby(["date"]).sum()["fare"]
# urban_fares = pyber_data_df.loc[(pyber_data_df["type"] == "Urban")].groupby(["date"]).sum()["fare"]

# type_fares_by_date = pd.DataFrame({
#                 "Rural":rural_fares,
#                 "Suburban":suburban_fares,
#                 "Urban":urban_fares,
#     })
# type_fares_by_date


fare_sum_by_type_date_pivot=fare_sum_by_type_date.pivot(index="date", columns="type", values="fare")
fare_sum_by_type_date_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
given_dates = fare_sum_by_type_date_pivot.loc['2019-01-01':'2019-04-29']
given_dates
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
given_dates.index= pd.to_datetime(given_dates.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
given_dates.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fare_sum=given_dates.resample('W').sum()
weekly_fare_sum
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')


weekly_fare_sum_plot = weekly_fare_sum.plot(figsize = (15,6))
# Add a title 
weekly_fare_sum_plot.set_title("Total Fare by City Type")
#  Add a x- and y-axis label.
weekly_fare_sum_plot.set_xlabel("")
weekly_fare_sum_plot.set_ylabel("Fare ($USD)")

plt.tight_layout()
plt.axis('tight')
plt.show()
```
